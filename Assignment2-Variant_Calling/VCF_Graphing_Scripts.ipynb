```python
import matplotlib.pyplot as plt


fs = open('year_vars_effect_pred.vcf', 'r')

counter = 0

DP_list = []
AF_list = []
QA_list = []
ANN_dict = {}

for line in fs:
    
    ##if counter <= 1000:
        
    if '#' not in line:
        linesplit = line.split(';')
        for searcher in linesplit:                

            if 'DP=' in searcher:
                if ',' not in searcher:
                    DP_list.append(float(searcher[3:]))

                else:
                    searcher_split = searcher.split(',')
                    for y in searcher_split:
                        if 'DP=' in y:
                            DP_list.append(float(y[3:]))
                        else:
                            DP_list.append(float(y))

            if 'AF=' in searcher:
                if 'SAF=' not in searcher:

                    if ',' not in searcher:
                        AF_list.append(float(searcher[3:]))

                    else:
                        searcher_split = searcher.split(',')
                        for z in searcher_split:
                            if 'AF=' in z:
                                AF_list.append(float(z[3:]))
                            else:
                                AF_list.append(float(z))

            if 'QA=' in searcher:
                if 'PQA=' not in searcher:

                    if ',' not in searcher:
                        QA_list.append(float(searcher[3:]))

                    else:
                        searcher_split = searcher.split(',')
                        for a in searcher_split:
                            if 'QA=' in a:
                                QA_list.append(float(a[3:]))
                            else:
                                QA_list.append(float(a))

            if 'ANN=' in searcher:
                split1 = searcher.split(',')
                for variation in split1:
                    if '|' in variation:
                        locstart = variation.index('|')
                        locend = variation[locstart+1:].index('|')
                        key_name = variation[locstart+1:locend+locstart+1]
                        ANN_dict.setdefault(key_name,0)
                        ANN_dict[key_name]+=1

##fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
##keys=ANN_dict.keys()
##values=ANN_dict.values()

##ax[0][0].bar(keys, values)
##ax[1][0].hist(DP_list)
##ax[0][1].hist(AF_list)
##ax[1][1].hist(QA_list)
##plt.show()

                            
                    ##searcher_split = searcher.split('|')
                    ##ANN_dict.setdefault(searcher_split[1],0)
                    ##ANN_dict[searcher_split[1]]+=1
                    
                                                
            ##counter += 1
                

```
